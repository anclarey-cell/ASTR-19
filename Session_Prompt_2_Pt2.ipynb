```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)

def decay_function(x, a, b):
    return a*np.exp(-b*x)
```

```python
a = 15
b = 7
```

```python
y = decay_function(x,a,b)
plt.figure(figsize=(8,5))
plt.plot(x, y, label = (f"y = {a} e^{{-{b}x}}"), color='blue', linewidth=2)

plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("X vs. Y", fontsize=14)
plt.legend()
plt.grid(True)
```

```python
yo_mama = np.linspace(2, 0.2, len(x))
New_yo = np.random.normal(loc=0.0, scale=yo_mama)

y_lmao = y + New_yo

plt.figure(figsize=(8,5))
plt.errorbar(x, y_lmao, yerr=yo_mama, fmt = 'o')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Data with Uncertainty")
plt.grid(True)
```

```python
plt.savefig("scatter_with_uncertainty.pdf", format='pdf')
plt.show()
```
