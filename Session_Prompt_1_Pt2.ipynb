```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)

def decay_function(x, a, b):
    return a*np.exp(-b*x)
```

```python
a = 15
b = 7
```

```python
y = decay_function(x,a,b)
plt.figure(figsize=(8,5))
plt.plot(x, y, label = (f"y = {a} e^{{-{b}x}}"), color='blue', linewidth=2)

plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("X vs. Y", fontsize=14)
plt.legend()
plt.grid(True)
```

```python
plt.savefig('decay_plot.pdf', format='pdf')
plt.show()
```
